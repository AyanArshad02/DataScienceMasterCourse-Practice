```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
y = np.cos(x)
```

```python
plt.plot(x,y)
```

```python
# labelling graph
plt.plot(x,y)
plt.xlabel("This is my X-axis data")
plt.ylabel("This is my Y-axis data")
plt.title("This is my title")
plt.show()
```

```python
"SCATTER PLOT"
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colour = np.random.rand(50)
sizes = 100*np.random.rand(50)
plt.scatter(x,y, c = colour, s = sizes) # we can change colour as well as size in scatter plot
```

```python
"Setting custom colour and size in scatter plot"
plt.scatter(x,y, c = "Blue", s = 50)
```

```python
"Changing Intensity of colour"
plt.scatter(x,y, alpha = 0.7)
```

```python
# SHOWING CATEGORICAL VS NUMERICAL DATA IN A BAR GRAPH 
x = ['a','b','c','d','e']
y = np.random.rand(5)
plt.bar(x,y)
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("This is my x-axis")
plt.ylabel("This is my y-axis")
plt.title("This is my title")
plt.show()
```

```python
# SHOWING BAR GRAPH IN HORIZONTAL MODE
plt.barh(x,y) # we will use barh instead of bar
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Horizontal Bar Graph")
plt.show()
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)

plt.figure(figsize=(8,3))
plt.grid() # grid aa jyega graph mein
plt.plot(x,y,'--y')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Title")
plt.show()
```

```python
# Now we want to see our Fare in INR. To do this we will make a new column and update fare
df['Fare_INR'] = df['Fare'].apply(lambda x : x*80) 
```

```python
def euro_to_inr(x):
    return x*80

df["Fare_INR"] = df['Fare'].apply(euro_to_inr)
```
